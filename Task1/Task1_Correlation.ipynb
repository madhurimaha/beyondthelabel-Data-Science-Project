# Naming issue Resolution


The names of the brands in the goodonyou dataset are the Parent Company or the Incorporated Name.

This section Merges the two datasets on the similar names.

Fuzzywuzzy is used for the string matching. It uses Levenshtein Distance to calculate the differences between sequences

```python
# code to show the mismatch of the name


merged = pd.merge(good,final_transparancy, on='Brands', how='outer', indicator=True)

# Filter for rows where the indicator column is 'both' (indicating a match)
matching_brands = merged[merged['_merge'] == 'both']

# Filter for rows where the indicator column is not 'both' (indicating a mismatch)
mismatching_brands = merged[merged['_merge'] != 'both']


print("Matching Brands:")
print(matching_brands[['Brands']])

print("\nMismatching Brands:")
print(mismatching_brands[['Brands']])
```

```python
from fuzzywuzzy import process,fuzz

df1 = good
df2 = final_transparancy


# Function to find the best match for each brand
def find_best_match(brand, choices):
    return process.extractOne(brand, choices, scorer=fuzz.token_set_ratio)


df1['BestMatch'] = df1['Brands'].apply(lambda x: find_best_match(x, df2['Brands'])[0])
df1['MatchScore'] = df1['Brands'].apply(lambda x: find_best_match(x, df2['Brands'])[1])
```

```python
# Merge the DataFrames based on the best match
merged_df_final = pd.merge(df1, df2, left_on='BestMatch', right_on='Brands', how='outer')

merged_df_final.head(20)
```

# Lasso

Lasso differs from traditional linear regression by introducing a regularization term called L1 regularization or "Lasso penalty" that forces some of the coefficients to be exactly zero. This results in a sparser model with fewer predictor variables, making it particularly useful for feature selection and model simplification.

The primary advantage of Lasso regression is its ability to perform automatic feature selection, identifying the most relevant variables while eliminating the less important ones. This not only improves model interpretability but also helps prevent overfitting by reducing the complexity of the model, ultimately leading to a better R-squared (R2) value and more robust predictions.

```python
from sklearn.linear_model import Lasso
# Initialize Lasso regression with default parameters
lasso_model = Lasso(alpha=1.0)
# Fit the Lasso model to the training data
lasso_model.fit(X_train, y_train)
# Make predictions on the test data
Y_pred = lasso_model.predict(X_test)
# Calculate the R-squared (R2) score
lasso_r2_score = r2_score(y_test, Y_pred)
print("Lasso Regression R-squared (R2) Score:", lasso_r2_score)
```

```python
# Initialize Lasso regression with specified parameters
alpha_value = 1e-3
max_iter_value = 100
tolerance_value = 0.01

lasso_model = Lasso(alpha=alpha_value, max_iter=max_iter_value, tol=tolerance_value)

# Fit the Lasso model to the training data
lasso_model.fit(X_train, y_train)


# Make predictions on the test data using the trained Lasso model
Y_pred = lasso_model.predict(X_test)

# Calculate the R-squared (R2) score
lasso_test_r2_score = r2_score(y_test, Y_pred)

print("Lasso Regression Test R-squared (R2) Score:", lasso_test_r2_score)
```

```python
# Write your code here
import random
from sklearn.model_selection import cross_val_score
alpha_arr = np.linspace(0.0001, 10, 10)

# Initialize an empty dictionary to store alpha and corresponding R2 scores
alpha_r2_scores = {}

for alpha in alpha_arr:

    ridge_model = Lasso(alpha=alpha)

    r2_scores = cross_val_score(ridge_model, X_train, y_train, cv=5, scoring='r2')

    # Calculate the average R2 score across cross-validation folds
    average_r2_score = sum(r2_scores) / len(r2_scores)

    # Store the alpha and its corresponding R2 score
    alpha_r2_scores[alpha] = average_r2_score

# Find the alpha with the highest R2 score
best_alpha = max(alpha_r2_scores, key=alpha_r2_scores.get)


print("Alpha with the Best R2 Score:", best_alpha)
print("R2 Score:", alpha_r2_scores[best_alpha])
```

# Ridge

Ridge regression introduces a regularization term known as L2 regularization or "Ridge penalty" that adds a penalty based on the square of the magnitude of coefficients. This encourages all coefficients to be small but not exactly zero.

The primary advantage of Ridge regression is its effectiveness in handling multicollinearity, a situation where predictor variables are highly correlated. By penalizing large coefficient values, Ridge regression can mitigate multicollinearity issues, leading to more stable and reliable coefficient estimates. This, in turn, results in better R-squared (R2) values and more robust model performance, particularly when dealing with datasets containing correlated features.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import RidgeCV
# Initialize a range of alpha values (logarithmic scale)
alphas = np.logspace(-4, 4, 10)
# Initialize RidgeCV with the range of alpha values and perform cross-validatio
ridge_cv = RidgeCV(alphas=alphas, store_cv_values=True)
ridge_cv.fit(X_train, y_train)
# Get the optimal alpha and corresponding R2 score
optimal_alpha = ridge_cv.alpha_
r2_score = np.max(np.mean(ridge_cv.cv_values_, axis=0))
print("Optimal Alpha:", optimal_alpha)
print("Cross-Validated R-squared (R2) Score:", r2_score)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge
# Define the range of alpha values (logarithmically spaced)
alphas = np.logspace(-2, 2, 10)
# Initialize lists to store alpha values and corresponding R2 scores
alpha_values = []
r2_scores = []
alpha_scores = []
# Perform cross-validation for each alpha value
for alpha in alphas:
    ridge_model = Ridge(alpha=alpha)
    r2 = np.mean(cross_val_score(ridge_model, X, y, cv=5, scoring='r2'))
    alpha_scores.append((alpha, np.mean(r2)))
    alpha_values.append(alpha)
    r2_scores.append(r2)
for alpha, score in alpha_scores:
    print(f"Alpha: {alpha}, R-squared (R2) Score: {score}")
# Find the alpha with the highest R2 score
best_alpha = alpha_values[np.argmax(r2_scores)]
best_r2_score = np.max(r2_scores)
print("Best Alpha:", best_alpha)
print("Best Cross-Validated R-squared (R2) Score:", best_r2_score)
# Plot the results
plt.figure(figsize=(10, 6))
plt.semilogx(alpha_values, r2_scores)
plt.xlabel('Alpha (log scale)')
plt.ylabel('Cross-Validated R2 Score')
plt.title('Cross-Validation Results for Ridge Regression')
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge
import numpy as np
import matplotlib.pyplot as plt

# Define the range of alpha values (logarithmically spaced)
alphas = np.logspace(-2, 2, 10)

# Initialize lists to store alpha values and corresponding MSE scores
alpha_values = []
mse_scores = []
alpha_scores = []

# Perform cross-validation for each alpha value
for alpha in alphas:
    ridge_model = Ridge(alpha=alpha)
    mse = -np.mean(cross_val_score(ridge_model, X, y, cv=5, scoring='neg_mean_squared_error'))
    alpha_scores.append((alpha, np.mean(mse)))
    alpha_values.append(alpha)
    mse_scores.append(mse)

for alpha, score in alpha_scores:
    print(f"Alpha: {alpha}, Mean Squared Error (MSE) Score: {score}")

# Find the alpha with the lowest MSE score
best_alpha = alpha_values[np.argmin(mse_scores)]
best_mse_score = np.min(mse_scores)
print("Best Alpha:", best_alpha)
print("Best Cross-Validated Mean Squared Error (MSE) Score:", best_mse_score)

# Plot the results
plt.figure(figsize=(10, 6))
plt.semilogx(alpha_values, mse_scores)
plt.xlabel('Alpha (log scale)')
plt.ylabel('Cross-Validated MSE Score')
plt.title('Cross-Validation Results for Ridge Regression')
plt.grid()
plt.show()
```

# Gradient Boosting Regressor

Gradient Boosting Regressor is an ensemble machine learning technique that builds a strong predictive model by sequentially adding decision trees to correct the errors of the previous ones. It starts with an initial prediction and then fits new trees to the residuals, continually improving accuracy. This iterative process minimizes prediction errors and creates a robust regression model.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd


# Create the model
gb_regressor = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Fit the model on the training data
gb_regressor.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = gb_regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

The Gradient Boosting Regressor, a powerful ensemble learning algorithm, was employed to predict sustainability scores based on transparency scores. This model excels in capturing complex relationships within the data by iteratively fitting decision trees, each correcting the errors of its predecessor. The low Mean Squared Error (MSE) of 0.2159 indicates the model's high precision, closely aligning predicted sustainability scores with actual values. Additionally, the R-squared value of 0.4993 suggests that of variability in sustainability scores can be explained by the transparency scores.

The model's  performance showcases its capability to discern subtle patterns in the data.

# Support Vector Regressor

Support Vector Regressor (SVR) works by finding a hyperplane in a high-dimensional space that best represents the relationship between input features and target values in a regression problem. Unlike traditional linear regression, SVR focuses on minimizing the margin of error within a predefined tolerance (epsilon) around the predicted values.

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import pandas as pd


# Standardize the features as SVR is sensitive to the scale of input features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


svr_regressor = SVR(kernel='linear', C=1.0)

# Fit the model on the training data
svr_regressor.fit(X_train_scaled, y_train)

# Make predictions on the testing data
y_pred = svr_regressor.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

The Support Vector Regressor (SVR), a machine learning algorithm designed for regression tasks, was employed to predict sustainability scores based on transparency scores.

SVR excels in capturing complex, non-linear relationships between variables by transforming the data into a higher-dimensional space.

In this context, the SVR achieved a Mean Squared Error (MSE) of 0.3150, indicating the average squared difference between predicted and actual sustainability scores.
The R-squared value of 0.2695 signifies the proportion of variability in sustainability scores explained by transparency scores.

While the SVR performance indicates a moderate fit to the data, its ability to handle non-linear relationships makes it a valuable tool for scenarios where linear models may fall short. The results suggest that, while the SVR captures some patterns, further exploration and refinement may enhance predictive accuracy.

# MLP Regressor

An MLP (Multi-Layer Perceptron) Regressor is a type of artificial neural network used for regression tasks. It consists of multiple layers of interconnected nodes, or neurons, each performing mathematical operations. During training, the network adjusts its weights and biases to minimize the difference between predicted and actual target values, gradually learning complex patterns and relationships in the data. This iterative process continues until the model achieves satisfactory predictive accuracy, making MLP Regressor a versatile tool for various regression problems.

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler


# Standardize the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize MLPRegressor
mlp_regressor = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)

# Fit the model
mlp_regressor.fit(X_train_scaled, y_train)

# Predictions
y_pred = mlp_regressor.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print results
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

 The model utilized a neural network architecture with one hidden layer containing 100 neurons and another hidden layer with 50 neurons. The mean squared error (MSE) of 0.267 indicates the average squared difference between predicted and actual values, providing a measure of the model's accuracy.

 The R-squared value of 0.3801 signifies the proportion of variance in the Brand Rating Numerical explained by the transparency scores, with a higher R-squared indicating a better fit.
