First the Scraping was done without punctuations and just the 'Parent Company'

Then the first word of the company was considered for the scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to convert a company name to the format used in the URL
def format_company_name(company_name):
    # Convert to lowercase and replace spaces with hyphens
    return company_name.lower().replace(" ", "-")

# Function to get the brand rating for a given company
def get_brand_rating(company_name):
    formatted_name = format_company_name(company_name)
    base_url = "https://directory.goodonyou.eco/brand/"
    url = base_url + formatted_name

    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        rating_element = soup.find('p', id='brand-rating')
        if rating_element:
            rating = rating_element.text.strip()
            return rating
        else:
            return "Rating not found"
    else:
        #print(f"Error for {company_name}: Status code {response.status_code}")
        return "Website not accessible"

# Read the list of company names from a text file
with open('/content/name_field.txt', 'r') as file:
    companies = [line.strip() for line in file]

# Initialize a dictionary to store company ratings
company_ratings = {'Company Name': [], 'Brand Rating': []}

# Iterate through the list of companies and fetch their brand ratings
for company in companies:
    brand_rating = get_brand_rating(company)
    company_ratings['Company Name'].append(company)
    company_ratings['Brand Rating'].append(brand_rating)

# Create a DataFrame from the dictionary
df = pd.DataFrame(company_ratings)

```
